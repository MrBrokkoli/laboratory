```python
import numpy as np
import scipy.integrate as integrate
import math
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
```

```python
result1 = integrate.quad(lambda x: 4 * math.sin(x) - 3, 0, 5)
result2 = integrate.dblquad(lambda x,y: x ** 2 + y ** 2, 0, 1, 0, 1)
print(result1)
print(result2)
```

```python
x = np.linspace(0, 20, num=100, endpoint=True)
y = (-np.sin(x)) ** 2
f = interp1d(x, y)
f2 = interp1d(x, y, kind='cubic')

xnew = np.linspace(0, 20, num=100, endpoint=True)
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```

```python
a = 27
b = 30

def f(x):
    return 1 / (a + b)

def F(x):
    return (x - a) / (b - a)

x = np.random.uniform(a, b, 20)
print(x)
plt.plot(x, f(x))
plt.plot(x, F(x))
plt.show()
```
