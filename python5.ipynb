```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math

x1 = [random.gauss(0, 1) for _ in range(30)]
x1 = x1[2:] + x1[:2]
print(x1)
x2 = [random.uniform(0, 3) for _ in range(30)]
x2 = x2[2:] + x2[:2]
print(x2)
```

```python
plt.figure(figsize = (8, 4))
plt.axis([1, 30, -2, 4])
plt.xticks(np.arange(1, 31))
plt.plot(range(1, 31), x1, range(1, 31), x2)
plt.legend(['x1', 'x2'])
plt.ylabel('Значения', rotation = 90)
plt.xlabel('Интервал', rotation = 0)
plt.show()
```
